```python
def load_and_process(url):
    df1 = (
        pd.read_csv(url)
        .dropna()
        .assign(GoalReachedPercentage = lambda x: x["pledged"] / x["goal"] * 100)
        .sort_values("GoalReachedPercentage", ascending = False)
        .reset_index(drop=True)
    )
    return df1
```

```python
df = load_and_process("ks-projects-201801.csv")
df
```
