```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#dataset before dropping NA values
df = pd.read_csv("kickstarter.csv")
df
```

```python
def load_and_process(url_or_path_to_csv_file):
    df1 = (
        pd.read_csv(url_or_path_to_csv_file)
        .dropna()
        .reset_index(drop=True)
    )
    df2 = (
        df1
        .assign(Percent_Funded= lambda x: x['pledged']/x['goal']*100)
        .reset_index(drop=True)
    )
    return df2

df = load_and_process("kickstarter.csv")
```

```python
#Dataset after removing 
df
```

```python
#Categories with the highest Mean Percent Funded listed in descending order
df2 = (
    df.loc[:,['main_category', 'Percent_Funded']]
    .groupby(by="main_category").mean()
    .sort_values('Percent_Funded', ascending=False)
    )
df2
```

```python
plt.figure(figsize=(25,10))
df2 = df[['main_category', 'Percent_Funded']]
sns.barplot(x = 'main_category', y='Percent_Funded', data=df2)
```

```python
# List of all failed projects
df3 = df[['name', 'main_category', 'goal', 'pledged', 'state', "Percent_Funded"]]
df3 = df3.loc[df3['state']=='failed'].sort_values('goal', ascending=False).reset_index(drop=True)
df3
```

```python
#List of failed projects, grouped by main_category, sorted by Percent_Funded in ascending order 
df4 = (
    df3.loc[:,['main_category', 'Percent_Funded']]
    .groupby(by="main_category").mean()
    .sort_values('Percent_Funded', ascending=True)
    )
df4
```
